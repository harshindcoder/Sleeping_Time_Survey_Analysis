# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Enhancements
sns.set(style='whitegrid')  # Modern grid style
```

```python
data1 = pd.read_csv('data/sleep_analysis_data.csv')
data2 = pd.read_csv('data/sleep_analysis_2_data.csv')
data = pd.concat([data1,data2])
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size

# Histogram with KDE
sns.histplot(data=data, x='Age', kde=True, color='royalblue', bins=20, alpha=0.7)

# Enhanced Titles and Labels
plt.title('Age Distribution of People in the Survey', fontsize=16, fontweight='bold', color='darkblue', pad=20)
plt.xlabel('Age of Survey Participants', fontsize=12, fontweight='medium', color='darkgreen', labelpad=10)
plt.ylabel('Number of People', fontsize=12, fontweight='medium', color='darkgreen', labelpad=10)

# Customizing Axis Ticks
plt.xticks(fontsize=10, color='darkred')
plt.yticks(fontsize=10, color='darkred')

# Add Grid Lines
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Display Plot
plt.tight_layout()  # Adjust spacing for clarity
plt.show()
```

Most of the people in the survey are young adults.

```python
plt.figure(figsize=(10, 6))  # Adjust figure size

# Bright Colors for the Countplot
sns.countplot(data=data, x='Gender', hue = 'Gender', palette='Set2')  # 'Set2' has bright pastel colors

# Enhanced Titles and Labels
plt.title('Gender Distribution of People in the Survey', fontsize=16, fontweight='bold', color='purple', pad=20)
plt.xlabel('Gender of Survey Participants', fontsize=12, fontweight='medium', color='blue', labelpad=10)
plt.ylabel('Number of People', fontsize=12, fontweight='medium', color='blue', labelpad=10)

# Customizing Axis Ticks
plt.xticks(fontsize=10, color='crimson')  # Bright red
plt.yticks(fontsize=10, color='crimson')

# Add Grid Lines
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Display Plot
plt.tight_layout()  # Adjust spacing for clarity
plt.show()
```

Amount of males and females are properly distributed in the survey so the data will not be biased for males and females but the results for other categories will be biased due to the small amount of data, we will try to avoid to make prediction for other categories.

```python
plt.figure(figsize=(10, 6))  # Adjust figure size

# Bright Colors for the Countplot
sns.countplot(data=data, x='meals/day', hue = 'meals/day', palette='Set2')  # 'Set2' has bright pastel colors

# Enhanced Titles and Labels
plt.title('Meals/day Distribution of People in the Survey', fontsize=16, fontweight='bold', color='purple', pad=20)
plt.xlabel('Meals/day of Survey Participants', fontsize=12, fontweight='medium', color='blue', labelpad=10)
plt.ylabel('Number of People', fontsize=12, fontweight='medium', color='blue', labelpad=10)

# Customizing Axis Ticks
plt.xticks(fontsize=10, color='crimson')  # Bright red
plt.yticks(fontsize=10, color='crimson')

# Add Grid Lines
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Display Plot

plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size

# Countplot with Bright Colors and Hue for Physical Illness
sns.countplot(data=data, x='physical illness', hue='physical illness', palette='Set2', dodge=True)

# Enhanced Titles and Labels
plt.title('Illness Issues of People in the Survey', fontsize=18, fontweight='bold', color='darkviolet', pad=20)
plt.xlabel('Illness Status of Survey Participants', fontsize=14, fontweight='medium', color='darkgreen', labelpad=12)
plt.ylabel('Number of People', fontsize=14, fontweight='medium', color='darkgreen', labelpad=12)

# Customizing Axis Ticks
plt.xticks(fontsize=12, color='darkred', rotation=0)  # Better readability with no rotation
plt.yticks(fontsize=12, color='darkred')

# Add Grid Lines with subtle styling
plt.grid(color='gray', linestyle='--', linewidth=0.7, alpha=0.5)

# Display Plot
plt.tight_layout()  # Adjust spacing for clarity
plt.show()
```

Most of the peoples participating in the survey are healthy so we can not make any comment on the peoples based on their health as the result would be biased.

```python
plt.figure(figsize=(10, 6))  # Adjust figure size

# Order on the x Axis
order_set = ['0-1 hrs', '1-2 hrs','2hrs', '2-3 hrs', '3-4 hrs', '4-5 hrs', 'more than 5']

# Countplot with Bright Colors and Hue for Physical Illness
sns.countplot(data=data, x='screen time', hue='screen time', order = order_set, palette='Set2')

# Enhanced Titles and Labels
plt.title('Screen time of People in the Survey', fontsize=18, fontweight='bold', color='darkviolet', pad=20)
plt.xlabel('Screen time of Survey Participants', fontsize=14, fontweight='medium', color='darkgreen')
plt.ylabel('Number of People', fontsize=14, fontweight='medium', color='darkgreen', labelpad=12)

# Customizing Axis Ticks
plt.xticks(fontsize=12, color='darkred')  # Better readability with no rotation
plt.yticks(fontsize=12, color='darkred')

# Add Grid Lines with subtle styling
plt.grid(color='gray', linestyle='--', linewidth=1, alpha=0.5)

# Display Plot
plt.tight_layout()  # Adjust spacing for clarity
plt.show()
```

Most of the people in the survey has screen time of more than two hours.
